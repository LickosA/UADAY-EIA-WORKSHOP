```python
import requests
import csv
import openpyxl

def extract_domain(url):
    """Extraction du nom de domaine d'une URL"""
    domain = url.split('//')[-1].split('/')[0]
    domain = domain.replace("www.", "")
    domain = domain.rstrip("/")
    return domain

def get_data(url):
    """Obtenir les données EIA et le nom du serveur d'une URL"""
    endpoint = "http://127.0.0.1:5000/mx_records/"
    domain = extract_domain(url)
    response = requests.get(endpoint+domain)
    if response.status_code == 200:
        eia = []
        server_name = []
        json_response = response.json()
        for data in json_response["mx_records"]:
            eia.append(data['isvalid_eia'])
            server_name.append(data['server_name'])
        return eia, server_name
    else:
        return None, None

# Liste des sites web
sites = []

wb = openpyxl.load_workbook('aau_members.xlsx')
feuille = wb['backup']
# Ajouter une nouvelle colonne
eia_colonne = feuille.max_column + 1
feuille.cell(row=1, column=eia_colonne, value='EIA')
server_colonne = feuille.max_column + 1
feuille.cell(row=1, column=server_colonne, value='SERVER NAME')
i=2
for row in feuille.iter_rows(min_row=2, max_row=feuille.max_row, values_only=True):
    print(i, row)
    if row[3] is not None:
       domain = extract_domain(row[3].strip())
       eia, server_name = get_data(domain)
       
    if eia:
        
        eia_str = ', '.join(str(e) for e in eia)
        print(eia_str)
        feuille.cell(row=i, column=6).value = eia_str
    if server_name:
        server_name_str = ', '.join(str(e) for e in server_name)
        print(server_name_str)
        feuille.cell(row=i, column=7).value = server_name_str
    i=i+1
    wb.save('aau_members.xlsx')
```
